```python
import models
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
M = models.MarkovChain(
    number_of_servers=1,
    arrival_rates=[5, 6],
    service_rates=[12, 13], 
    class_change_rate_matrix=[
        [None, 3],
        [1, None]
    ],
    inf=40)
```

```python
S = models.Simulation(
    number_of_servers=1,
    arrival_rates=[5, 6],
    service_rates=[12, 13], 
    class_change_rate_matrix=[
        [None, 3],
        [1, None]
    ],
    max_simulation_time=10000,
    warmup=300
)
```

```python
states = range(50)
```

```python
fig, ax = plt.subplots(1)
ax.plot(states, [M.aggregate_probs.get(s, 0) for s in states], label='Markov Chain')
ax.plot(states, [S.aggregate_probs.get(s, 0) for s in states], label='Simulation')
ax.set_xlabel('Number of Customers', fontsize=14)
ax.set_ylabel('Probability', fontsize=14)
ax.legend(fontsize=14, frameon=True)
plt.show()
```
