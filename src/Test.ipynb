```python
import models
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

# Test for 2 classes (more likely to transition to hgiher class)

```python
M = models.MarkovChain(
    number_of_servers=1,
    arrival_rates=[5, 6],
    service_rates=[12, 13], 
    class_change_rate_matrix=[
        [None, 3],
        [1, None]
    ],
    inf=40)
```

```python
S = models.Simulation(
    number_of_servers=1,
    arrival_rates=[5, 6],
    service_rates=[12, 13], 
    class_change_rate_matrix=[
        [None, 3],
        [1, None]
    ],
    preempt=True,
    max_simulation_time=20000,
    warmup=200
)
```

```python
fig, axarr = plt.subplots(1, 3, figsize=(17, 5))
states = range(50)
axarr[0].plot(states, [M.aggregate_probs.get(s, 0) for s in states], label='Markov Chain')
axarr[0].plot(states, [S.aggregate_probs.get(s, 0) for s in states], label='Simulation')
axarr[0].set_xlabel('Number of Customers', fontsize=14)
axarr[0].set_ylabel('Probability', fontsize=14)
axarr[0].legend(fontsize=14, frameon=True)

for clss in range(M.k):
    axarr[1].plot(states, [M.aggregate_probs_by_class[clss].get(s, 0) for s in states], label=f"Class {clss}")
axarr[1].set_title('Markov Chain', fontsize=16)
axarr[1].legend(fontsize=14, frameon=True)
axarr[1].set_xlabel('Number of Customers', fontsize=14)
axarr[1].set_ylabel('Probability', fontsize=14)

for clss in range(S.k):
    axarr[2].plot(states, [S.aggregate_probs_by_class[clss].get(s, 0) for s in states], label=f"Class {clss}")
axarr[2].set_title('Simulation', fontsize=16)
axarr[2].legend(fontsize=14, frameon=True)
axarr[2].set_xlabel('Number of Customers', fontsize=14)
axarr[2].set_ylabel('Probability', fontsize=14)

plt.show()
```
