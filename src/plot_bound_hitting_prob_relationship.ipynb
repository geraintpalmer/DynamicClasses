```python
import models
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("experiments/bound_hitting_prob_data.csv")
```

```python
data_A = data[data.Example == "A"].sort_values("Bound")
data_B = data[data.Example == "B"].sort_values("Bound")
data_C = data[data.Example == "C"].sort_values("Bound")
data_D = data[data.Example == "D"].sort_values("Bound")
```

```python
fig, ax = plt.subplots(1)
ax.plot(
    data_A["Bound"], data_A["Max_hitting_prob"], c="black", linestyle="solid", label="A"
)
ax.plot(
    data_B["Bound"],
    data_B["Max_hitting_prob"],
    c="black",
    linestyle="dashed",
    label="B",
)
ax.plot(
    data_C["Bound"],
    data_C["Max_hitting_prob"],
    c="black",
    linestyle="dotted",
    label="C",
)
ax.plot(
    data_D["Bound"],
    data_D["Max_hitting_prob"],
    c="black",
    linestyle="dashdot",
    label="D",
)
ax.legend(fontsize=12, frameon=True, title="Example", title_fontsize=14)
ax.set_xlabel("Bound", fontsize=14)
ax.set_ylabel(
    "Maximum Probability of Hitting the\n Boundary from Reasonable Region", fontsize=12
)
ax.set_ylim(0, 0.225)
ax.set_yticks([0, 0.05, 0.1, 0.15, 0.2])
ax.set_xlim(6, 30)
ax.set_title("Reasonable Region 80% of Bound")
fig.savefig("../tex/img/bound_hitting_prob_relationship.pdf")
```
